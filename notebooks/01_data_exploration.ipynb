# 01 — Data Exploration

Explore the **WebUOT-238-Test** dataset from HuggingFace — a 238-video subset of the WebUOT-1M benchmark (NeurIPS 2024).

This notebook inspects the annotation structure that will be used in notebook 02 to generate VQA training pairs:
- Video clips and metadata
- Language prompts (target descriptions)
- Bounding box annotations
- 23 tracking attributes (low visibility, camouflage, fast motion, etc.)
- Environment and superclass labels

**Dataset:** `Voxel51/WebUOT-238-Test`  
**Next step:** `02_vqa_generation.ipynb`

```python
!pip install -U fiftyone -q

import fiftyone as fo
from fiftyone.utils.huggingface import load_from_hub

dataset = load_from_hub("Voxel51/WebUOT-238-Test")
print(dataset)
```

```python
sample = dataset.first()

# Language prompt — the seed for QA generation
print("Language prompt:", sample.language)

# All 23 attributes and their values
attributes = [
    "Low Resolution", "Fast Motion", "Scale Variations", "Aspect Ratio Variations",
    "Camera Motion", "Viewpoint Changes", "Partial Occlusion", "Full Occlusion",
    "Out-of-View", "Rotation", "Deformation", "Similar Distractors",
    "Illumination Variations", "Motion Blur", "Partial Target Information",
    "Natural or Artificial Object", "Camouflage", "Underwater Visibility",
    "Watercolor Variations", "Underwater Scenarios", "Shooting Perspective",
    "Size", "Length"
]

print("\n--- ATTRIBUTES ---")
for attr in attributes:
    val = getattr(sample, attr.replace(" ", "_") if hasattr(sample, attr.replace(" ", "_")) else attr, None)
    # FiftyOne uses the field name as-is, try bracket access
    try:
        val = sample[attr]
        print(f"{attr}: {val.label if val else None}")
    except:
        print(f"{attr}: (access error)")

# Bounding boxes — check first 3 frames
print("\n--- FRAME BBOXES (first 3 frames) ---")
for frame_num, frame in list(sample.frames.items())[:3]:
    print(f"Frame {frame_num}: {frame.gt}")

# Video filepath
print("\n--- VIDEO PATH ---")
print(sample.filepath)

# Check a few more samples for language prompt variety
print("\n--- LANGUAGE PROMPTS (first 5 samples) ---")
for s in dataset.take(5):
    print(f"  {s.language}")
```
